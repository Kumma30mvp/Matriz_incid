# Matriz de incidencia

```python
# Modulos
import nltk 
nltk.download('punkt')
```

## 1- Definir el stoplist

```python
with open("stoplist.txt") as file:
    stoplist = [line.strip() for line in file]
stoplist += ['?', '-', '.', ':', ',']    
```

## 2-  Tokenizar y eliminar stopwords

```python

list_docs = []
for i in range(1, 7):
    libro = f"libro{i}.txt"
    with open(libro) as file:
        doc = [line.strip() for line in file]
        list_docs.append(" ".join(doc)) 

for i, doc in enumerate(list_docs):
    tokens = nltk.word_tokenize(doc)  
    tokens = [word for word in tokens if word.lower() not in stoplist]  
    list_docs[i] = tokens  
    print(f"Libro {i+1}: {tokens}")  

```

## 3- Construir la matriz de incidencia

```python
import numpy as np
def construir_matriz(list_docs):
    vocabulario = set()
    for doc in list_docs:
        vocabulario.update(doc)
    matriz = np.zeros((len(list_docs), len(vocabulario)), dtype=int)

    for i, doc in enumerate(list_docs):
        for palabra in doc:
            j = list(vocabulario).index(palabra)
            matriz[i][j] += 1
    
    return matriz, vocabulario
matriz_incidencia, vocabulario = construir_matriz(list_docs)
with open("matriz_incidencia.txt", "w") as file:
    file.write("\t".join(vocabulario) + "\n")
    
    for fila in matriz_incidencia:
        file.write("\t".join(map(str, fila)) + "\n")
        print(fila)
        print("\n")
```

## 4- Aplicar consultas booleanas

```python
import numpy as np

def indice_palabra(word, vocabulario):
    return list(vocabulario).index(word)

def M(word, vocabulario):
    columna = np.zeros((len(vocabulario)), dtype=int)
    columna[indice_palabra(word, vocabulario)] = 1
    return columna

def AND(A, B):
    return np.logical_and(A, B).astype(int)

if "frodo" in vocabulario:
    Query1 = AND(M("anillo", vocabulario), M("frodo", vocabulario))
    print("anillo AND frodo: ", Query1)
else:
    print("Word 'frodo' is not in the vocabulary.")
```
